```python
import subprocess
import os
```

```python
# Base directory
base_directory = "/media/toor/T7Shield/Mouse1637/Linear"
```

```python
# Directory to change to before running the bash script
bash_working_directory = "/home/toor/Desktop/stability-preprocessing/"
os.chdir(bash_working_directory)
```

### Using preproc_session.sh

```python
# Walk through the directory and subdirectories
# count = 0
for root, dirs, files in os.walk(base_directory):
    # Initialize file paths
    miniscope_file = None
    behavior_linear_file = None
    behavior_file = None

    # Look for the required files in the current directory
    for file in files:
        if "miniscope" in file:
            miniscope_file = os.path.join(root, file)
            print(str(miniscope_file))
        elif "behaviorLinear" in file:
            behavior_linear_file = os.path.join(root, file)
            print(str(behavior_linear_file))
        elif file.startswith("behavior") and not file.startswith("trimmed_behaviorLinear"):
            behavior_file = os.path.join(root, file)
            print(str(behavior_file))

    # Check if all required files are found
    if miniscope_file and behavior_linear_file and behavior_file:
        # Construct and run the bash command
        bash_command = [
            "bash", 
            "scripts/preproc_session.sh", 
            miniscope_file, 
            behavior_linear_file, 
            behavior_file
        ]
        print(f"Running command: {' '.join(bash_command)}")
        try:
            subprocess.run(bash_command, check=True)
        except subprocess.CalledProcessError as e:
            print(f"Error while processing {root}: {e}")

    # count += 1
    # if count == 2:
    #     break
```

### Using preproc_caiman.py

```python
# Walk through the directory and subdirectories
for root, dirs, files in os.walk(base_directory):
    # Initialize file paths
    miniscope_file = None
    motion_correction = None

    # Look for the required files in the current directory
    for file in files:
        if "miniscope" in file:
            miniscope_file = os.path.join(root, file)
            
        if "motion_correction_comparison" in file: # this means that the file is already processed
            motion_correction = os.path.join(root, file)

        # Check if all required files are found
    if miniscope_file and not motion_correction:
        # use os.path.expanduser() to handle the home directory
        script_path = os.path.expanduser("~/Desktop/stability-preprocessing/scripts/preproc_caiman.py")
        
        # Construct and run the bash command
        bash_command = [
            "python", 
            script_path,
            "--input_path", 
            miniscope_file
            ]
        print(f"Running command: {' '.join(bash_command)}")
        try:
            subprocess.run(bash_command, check=True)
        except subprocess.CalledProcessError as e:
            print(f"Error while processing {root}: {e}")

```
